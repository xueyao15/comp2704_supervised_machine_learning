### Executive Summary

This project aims to develop a model to predict NHL skatersâ€™ salaries based on a skater dataset of 2016/2017 season. The model can guide funding allocations and increase funding efficiency for teams. During data cleaning, 10 irrelevant columns and 38 rows with missing values are dropped and 390 missing values are filled using different strategies. The analysis shows that the salary distribution is highly right skewed and some player information (eg. Age) and performance statistics (eg. TOI) demonstrate to be better predictors of the salary. A regression algorithm will be applied to predict a continuous value (salary). (Word count: 98)

### Use Case

Limited by the salary cap, each NHL team attaches great attention to the allocation of their funding in order to recruit more skilled players at a low cost. This model reveals the statistical information that most affects salary and provides a benchmark value for team managers to monitor the dynamic value change of each player and assign a reasonable salary allocation to their players. With the help of this model, it will greatly increase the funding efficiency and strengthen our team. Especially during NHL trades, the results of such a prediction could be leveraged when more consideration is given to trading overpaid players and keeping underpaid ones. On the other hand, couch can motivate players for further improvement with those statistics that closely relate to salary levels.

### Data Selection and Cleaning

This dataset features the salaries of 888 NHL players for the 2016/2017 season. There are 143 columns with playersâ€™ bio information and performance metrics as well as one column with players 2016/2017 annual salary. 
This dataset was created by Rob Vollman and acquired from his [Hockey Abstract](http://www.hockeyabstract.com/testimonials/nhl2016-17playerdata) website.

```python
# Importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Displaying all columns and rows
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows', None)
```

```python
# Loading data
df = pd.read_csv('nhl_player_16_17.csv', sep=',').copy()
```

```python
# Displaying the first five rows
df.head()
```

```python
# Dispalying dataset information
df.info()
```

```python
# Checking if there is any duplicate values in dataset.
df.duplicated().sum()
```

There is no duplicate value in our dataset.

By examining the dataset, we found some bio information including City, Province or state of birth, Nationality, Ht, Wt, Hand, Last name, First Name, Team, NHLID is irrelevant to predicting salary, as well as the shoutout metrics, including Shootout Shots, Shootout Goals, Game-deciding shootout goals.

```python
# Dropping irrelevant features
df.drop(['City','Pr/St','Nat','Ht','Wt','Hand','Last Name', 'First Name','Team','NHLid', 'SOS','SOG','SOGDG'],inplace=True,axis=1)
df.head()
```

```python
# Checking each column's datatype.
df.dtypes
```

```python
# Converting the datatype of IPP%(individual points percentage) from object to float.
df['IPP%'] = pd.to_numeric(df['IPP%'], errors='coerce')
df.dtypes
```

```python
# Checking missing values.
df.isna().sum()
```

```python
# We can see that 130 players went undrafted so we will use the max draft number +1 to fill undrafted players' overall draft picks.
draft_ovrl = df['Ovrl'].max()+1
df['Ovrl'] = df['Ovrl'].fillna(draft_ovrl)
df['Ovrl']
```

```python
# Similary to Ovrl, we use the max draft number +1 to fill undrafted players' draft rounds.
draft_round = df['DftRd'].max()+1
df['DftRd'] = df['DftRd'].fillna(draft_round)
df['DftRd']
```

```python
# Filling the missing draft years with the mean of the DftYr column.
draft_year = df['DftYr'].mean()
df['DftYr'] = df['DftYr'].fillna(draft_year).astype(int)
df['DftYr']
```

```python
# Dropping rows with missing data and checking to make sure we have no more missing values
dropna = ['Salary','TOI%','IPP%','SH%','Pass','GS/G','iFF','iHA']
df.dropna(subset=dropna, inplace = True)
df.isna().sum()
```

There is no missing data in our dataset.

```python
# Adding age column and roughly calculating age based off birth year and dropping Born column. *note player data is from 2016-2017 season
df['Born'] = pd.DatetimeIndex(df['Born']).year
df['Age'] = 2016 - df['Born']
df = df.drop(columns=['Born'])
df['Age']
```

```python
# Checking how many unique values in Position column.
df['Position'].unique()
```

```python
# Seperating the four positions and adding columns for each position.
df['LW'] = df['Position'].str.findall('LW')
df['RW'] = df['Position'].str.findall('RW')
df['D'] = df['Position'].str.findall('D')
df['C'] = df['Position'].str.findall('C')
df.head()
```

```python
# Assigning 0 to empty lists and 1 to the rest for four position columns and dropping Position column.
df['LW'] = df['LW'].apply(lambda x: 0 if x==[] else 1)
df['RW'] = df['RW'].apply(lambda x: 0 if x==[] else 1)
df['D'] = df['D'].apply(lambda x: 0 if x==[] else 1)
df['C'] = df['C'].apply(lambda x: 0 if x==[] else 1)
df = df.drop(columns=['Position'])
df[['LW','RW','D','C']]
```

```python
# An overview of our dataset after cleaning
df.info()
```

```python
# Plotting the distribution of salaries in different ages
plt.figure(figsize = (10,6))
sns.boxplot(x='Age',y='Salary',data=df)
```

It is reasonable to assume that age has some influence on the salary of a player. Younger players may lack considerable experience within the NHL, so their contracts are valued lower since they are not expected to play many games in the NHL. Older players, on the other hand, may be more likely to have injuries as a result of a long-term career in the NHL and downhill physical performance, so a lower salary is often expected. It is worth to note that salary begins to show disparity around age 22 when players are finishing their rookie contracts and talented players with more expectations will sign a contract with a soaring salary. It is also interesting to see that all outliers (players earning much more than their age group) and the maximum salary earner appear before 30. That could be explained by players maximizing the expectations plus their performance before they are past their prime. 

```python
# Calculating the mean and median salaries for each position
lw = df.groupby('LW').get_group(1)
lw_salary = lw['Salary'].agg([np.mean, np.median])
lw_df = pd.DataFrame(lw_salary)

rw = df.groupby('RW').get_group(1)
rw_salary = rw['Salary'].agg([np.mean, np.median])
rw_df = pd.DataFrame(rw_salary)

d = df.groupby('D').get_group(1)
d_salary = d['Salary'].agg([np.mean, np.median])
d_df = pd.DataFrame(d_salary)

c = df.groupby('C').get_group(1)
c_salary = c['Salary'].agg([np.mean, np.median])
c_df = pd.DataFrame(c_salary)

position_df = pd.concat([lw_df,rw_df,d_df,c_df])

position_df = position_df.T
position_df.columns=['lw_mean','lw_median','rw_mean','rw_median','d_mean','d_median','c_mean','c_median']
position_df
```

```python
# Plotting the mean and median salaries for each position
plt.figure(figsize = (10,6))
sns.barplot( data = position_df)
```

After pulling out features that show a weak correlation with salary, we further examed these features as below:

1.+/- , SV%, A/60, FO%

we found that metrics like +/-  (a cumulative total goals scored for and against a playerâ€™s team while the player is on the ice), SV% (the team's save percentage when the player was on the ice) and A/60 (shots attempts by opposition per 60 minutes), FO% (faceoff winning percentage)
 are more related to a teamâ€™s performance. We may drop these features at the training phase.

2.BLK%, iHDf

The reason that BLK% (Â Percentage of all opposing shot attempts blocked by this player) shows a weak correlation with salary may be due to the position of the player takes in the game, for example, a good forward may never have a higher BLK% than a good defender but they could both enjoy a high salary. Same logic applies to iHDf (the difference in hits thrown by this individual minus those taken) as well; the value varies depending on the different roles assigned to players in the team.

3.PSG, iPenDf, Maj, Match, Misc, Game

PSG (penalty shot goals), iPenDf (the difference in penalties drawn minus those taken), Maj(Major penalties taken), Match(Match penalties), Misc(Misconduct penalties), Game(Game Misconduct penalties) are all penalty metrics that do not show much correlation with salary.

4.sDist, DAP, %FOT

DAP(disciplined aggression proxy) which is hits and takeaways divided by minor penalties, sDist (the average shot distance of shots taken by this player) and %FOT (Percentage of all on-ice faceoffs taken by this player.) are not that relevant with how much a player can earn. A fun fact to know here is DAP which measures how good a player is at getting the puck away from the opponents does not seem to weight much when it comes to salary.

Though huge income disparity between mean and median salaries in each position is shown above, the salaries of the right wingers seem to excel in both the mean and the median salaries compared with other positions. In stark contrast, the mean and median salaries of left wingers are at bottom.
